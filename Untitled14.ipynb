```python
print("user what do want do")
tasks=[]
while True:
  print("1,add task")
  print("2,view task")
  print("3,remove task")
  print("4,exit")

  choice=int(input("enter the number b/w (1-4)"))
  if choice==1:
   task=(input("add new tasks:"))
   tasks.append(task)
   print("task add sucessful")
  elif choice==2:
    if len(tasks)==0:
     print("no tasks availble")
    else:
     print("your task")
     for i in range(len(tasks)):
      print(f"{i+1}.{tasks[i]}")
  elif choice==3:
    if len(tasks)==0:
      print("no tasks delete")
    else:
      for i in range(len(tasks)):
        print(f"{i+1}.{tasks[i]}")
        delete_tasks=int(input("enter task number to delete"))
        if delete_tasks>0 and delete_tasks<=len(tasks):
           remove=(tasks.pop(delete_tasks-1))
           print(f"removed task:{remove}")
  elif choice==4:
   print("exit app")
   break
  else:
   print("invalid choice")
```
